```python
def get_number(input_txt):
    while True:
        try:
            user_input = input(input_txt).strip()
            if user_input.lower() == 'q':
                return 'quit'
            return float(user_input)
        except ValueError:
            print("Invalid input! Please enter a valid number (or 'q' to quit).")

def get_operator():
    valid_operators = ['+', '-', '*', '/']
    while True:
        op = input("Choose an operator (+, -, *, /) or 'q' to quit: ").strip()
        if op.lower() == 'q':
            return 'quit'
        if op in valid_operators:
            return op
        print(f"Invalid operator! Please choose from: {', '.join(valid_operators)}")

def calculator():
    print("Welcome to the Mini Calculator!")
    print("Type 'q' to quit at any prompt to exit.\n")
    
    while True:
        print("-" * 40)
        
        num1 = get_number("Enter the first number: ")
        if num1 == 'quit':
            break
            
        operator = get_operator()
        if operator == 'quit':
            break
            
        num2 = get_number("Enter the second number: ")
        if num2 == 'quit':
            break
        result = None
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("Math Error: Cannot divide by zero!")
                continue
            result = num1 / num2

        if result is not None:
            if result.is_integer():
                result = int(result)
            print(f" Result: {num1} {operator} {num2} = {result}")
            
    print("\nThank you for using Mini Calculator. Goodbye!")

if __name__ == "__main__":
    calculator()
```
